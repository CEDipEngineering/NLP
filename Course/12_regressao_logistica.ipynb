```python
df = pd.read_csv('./datasets/spam_ham_dataset.csv')
df.head()
```

```python
textos = list(df['text'])
labels = list(df['label'])
vectorizer = CountVectorizer(binary=True)
X = vectorizer.fit_transform(textos)
X_train, X_test, y_train, y_test = train_test_split(X, labels)
```

```python
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
y_pred = model_lr.predict(X_test)
print(classification_report(y_test, y_pred))
print(model_lr.classes_)
```

# Exercício 3
*Objetivo: comparar a relevância de palavras em regressão logística com a encontrada em Naive Bayes*

Nos códigos abaixo, temos uma ideia de como encontrar a relevância de palavras na Regressão Logísitica.

1. Como ela se compara com aquela fornecida usando a frequência de palavras em Naive Bayes?
1. Como ela se compara com aquela fornecida usando a métrica de Chen et al., que implementamos na aula passada, para o caso Naive Bayes?

```python
vocab = vectorizer.vocabulary_
print(model_lr.classes_)
palavras_ham = []
palavras_spam = []
for t in vocab.keys():
    coef = model_lr.coef_[0, vocab[t]]
    if coef<0:
        palavras_ham.append( (-coef, t) )
    else:
        palavras_spam.append( (coef, t) )
```

```python
tuplas_ordenadas = sorted(palavras_ham, reverse=True) # reverse=True pede uma ordenação em ordem decrescente
palavras = [ t[1] for t in tuplas_ordenadas ]
contagens = [ t[0] for t in tuplas_ordenadas ]

n_palavras = 30
eixo_x = np.arange(n_palavras)
plt.figure(figsize=(14,3))
plt.bar(eixo_x[0:n_palavras], contagens[0:n_palavras])
plt.xticks(eixo_x[0:n_palavras], palavras[0:n_palavras], rotation=90)
plt.show()
```

```python
tuplas_ordenadas = sorted(palavras_spam, reverse=True) # reverse=True pede uma ordenação em ordem decrescente
palavras = [ t[1] for t in tuplas_ordenadas ]
contagens = [ t[0] for t in tuplas_ordenadas ]

n_palavras = 30
eixo_x = np.arange(n_palavras)
plt.figure(figsize=(14,3))
plt.bar(eixo_x[0:n_palavras], contagens[0:n_palavras])
plt.xticks(eixo_x[0:n_palavras], palavras[0:n_palavras], rotation=90)
plt.show()
```

```python
# Resolva aqui
def analyse_dataset(path_to_csv: str, n_models = 100, text_col = 'text', label_col = 'label'):
    df = pd.read_csv(path_to_csv)
    textos = list(df[text_col])
    labels = list(df[label_col])
    vectorizer = CountVectorizer(binary=True)
    X = vectorizer.fit_transform(textos)

    example_label = df[label_col][0]

    scores = []
    for _ in range(n_models):
        X_train, X_test, y_train, y_test = train_test_split(X, labels)

        model_nb = BernoulliNB()
        model_nb.fit(X_train, y_train)
        y_pred = model_nb.predict(X_test)
        score_nb = classification_report(y_test, y_pred, output_dict=True)[example_label]["f1-score"]
        model_lr = LogisticRegression(max_iter=1000)
        model_lr.fit(X_train, y_train)
        y_pred = model_lr.predict(X_test)
        score_lr = classification_report(y_test, y_pred, output_dict=True)[example_label]["f1-score"]

        scores.append([score_nb, score_lr])

    scores_nb_distro = [x[0] for x in scores]
    scores_lr_distro = [x[1] for x in scores]
    nb_mean = np.mean(scores_nb_distro)
    lr_mean = np.mean(scores_lr_distro)
    nb_quant = [x - nb_mean for x in np.quantile(scores_nb_distro, (0.05, 0.95))]
    lr_quant = [x - lr_mean for x in np.quantile(scores_lr_distro, (0.05, 0.95))]
    x_label=["NaiveBayes", "LogReg"]
    plt.figure(figsize=(4,9))
    plt.bar(np.arange(len(x_label)), [nb_mean, lr_mean], yerr=[nb_quant, lr_quant])
    plt.xticks(np.arange(len(x_label)), x_label, rotation=90)
    plt.show()
```

```python
analyse_dataset('./datasets/spam_ham_dataset.csv')
```
